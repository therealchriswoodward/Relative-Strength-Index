```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = range(1,51)
y = [1,2,3,4,5,20,26,50,75,90,80,75,70,30,20,11,9,5,2,1,
     10,11,13.5,15,26,35,39,59,68,40,11,99,26,34,33,30,25,49,45,50,
     51,53,56,54,59,61,62,65,63,64]
```

```python
df = pd.DataFrame({
    'x': x,
    'y': y
})
```

```python
# RSI Oscillator

delta = df['y'].diff(1)
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
rsi = 100 - (100 / (1 + rs))

df['RSI'] = rsi
```

```python
fig, axs = plt.subplots(2,1)

axs[0].plot(x, y, color='black', label='Data')
axs[0].set_title("This is the data")
axs[0].set_xlim(0,50)

axs[1].plot(x,df['RSI'], color='red')
axs[1].axhline(70,color='black', linestyle="--")
axs[1].axhline(30,color='black', linestyle="--")
axs[1].set_title("RSI Oscillator")
axs[1].set_xlim(0,50)
axs[1].set_ylim(0,103)

plt.tight_layout()
```
